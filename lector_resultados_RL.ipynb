```python
import pybullet as p, pybullet_data
from Archivos_Apoyo.Pybullet_Robot_Data import PyBullet_Robot_Data
client=p.connect(p.DIRECT)
p.setAdditionalSearchPath(pybullet_data.getDataPath())
plane = p.loadURDF("plane.urdf")
robot_list=("2_legged_human_like_robot20DOF","2_legged_human_like_robot12DOF","2_legged_human_like_robot12DOF_done")
robot_name=robot_list[2]
robot = p.loadURDF(f"Robots_Versiones/{robot_name}.urdf", [0,0,1.21], useFixedBase=True)
num_joints=p.getNumJoints(robot)
informacion_robot=PyBullet_Robot_Data(robot)
joint_info=informacion_robot._get_joint_info

link_info = informacion_robot._get_link_info
p.disconnect(client)
```

```python
from json import dumps
json_joint_info=dumps(joint_info, indent=4)
json_link_info=dumps(link_info,indent=4)
with open(f"JSON_Info_Robot/{robot_name}.json","w") as f:
    f.write(json_joint_info)
with open(f"JSON_Info_Robot/{robot_name}_links.json","w") as f:
    f.write(json_link_info)
```

```python
from json import load
with open(f"JSON_Info_Robot/{robot_name}.json","r") as f:
    json_file=load(f)
with open(f"JSON_Info_Robot/{robot_name}_links.json","r") as f:
    json_file_link=load(f)
```
