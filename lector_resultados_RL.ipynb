```python
import pybullet as p, pybullet_data, time
p.connect(p.GUI)
p.setAdditionalSearchPath(pybullet_data.getDataPath())
plane = p.loadURDF("plane.urdf")
robot = p.loadURDF("2_legged_human_like_robot.urdf", [0,0,1.2], useFixedBase=True)

# desactiva motores por defecto en todos los joints
for j in range(p.getNumJoints(robot)):
    p.setJointMotorControl2(robot, j, p.VELOCITY_CONTROL, force=0)

LHIP, RHIP = 0, 3
LKNEE,RKNEE = 1,4
for t in range(4000):
    p.setJointMotorControl2(robot, LHIP, p.TORQUE_CONTROL, force=-50.0)  # SOLO izquierda
    p.setJointMotorControl2(robot, LKNEE, p.TORQUE_CONTROL, force=50.0)  # SOLO izquierda
    p.setJointMotorControl2(robot, RHIP, p.TORQUE_CONTROL, force=0.0)   # derecha a 0
    p.stepSimulation()
    if t%40==0:
        print("τR=", p.getJointState(robot, RHIP)[3])
        print("τL=", p.getJointState(robot, LHIP)[3])
```
