```python
import pybullet as p, pybullet_data, time
p.connect(p.GUI)
p.setAdditionalSearchPath(pybullet_data.getDataPath())
plane = p.loadURDF("plane.urdf")
robot = p.loadURDF("2_legged_human_like_robot.urdf", [0,0,1.2], useFixedBase=True)

# desactiva motores por defecto en todos los joints
for j in range(p.getNumJoints(robot)):
    p.setJointMotorControl2(robot, j, p.VELOCITY_CONTROL, force=0)

LHIP, RHIP = 0, 3
LKNEE,RKNEE = 1,4
for t in range(4000):
    p.setJointMotorControl2(robot, LHIP, p.TORQUE_CONTROL, force=-50.0)  # SOLO izquierda
    p.setJointMotorControl2(robot, LKNEE, p.TORQUE_CONTROL, force=50.0)  # SOLO izquierda
    p.setJointMotorControl2(robot, RHIP, p.TORQUE_CONTROL, force=0.0)   # derecha a 0
    p.stepSimulation()
    if t%40==0:
        print("τR=", p.getJointState(robot, RHIP)[3])
        print("τL=", p.getJointState(robot, LHIP)[3])
```

```python
import pybullet as p, pybullet_data
client=p.connect(p.DIRECT)
p.setAdditionalSearchPath(pybullet_data.getDataPath())
plane = p.loadURDF("plane.urdf")
robot = p.loadURDF("2_legged_human_like_robot12DOF.urdf", [0,0,1.21], useFixedBase=True)
num_joints=p.getNumJoints(robot)

for joint_id in range(num_joints):
    info_joint=p.getJointInfo(robot, joint_id)
    print()
    print("Index", info_joint[0])
    print("Name", info_joint[1])
    print("jointType", info_joint[2])
    print("qIndex", info_joint[3])
    print("uIndex", info_joint[4])
    print("flags", info_joint[5])
    print("jointDamping", info_joint[6])
    print("jointFriction", info_joint[7])
    print("jointLowerLimit", info_joint[8])
    print("jointUpperLimit", info_joint[9])
    print("jointMaxForce", info_joint[10])
    print("jointMaxVelocity", info_joint[11])
    print("linkName", info_joint[12])
    print("jointAxis", info_joint[13])
    print("parentFramePos", info_joint[14])
    print("parentFrameOrn", info_joint[15])
    print("parentFrameOrn", info_joint[16])





p.disconnect(client)
```
