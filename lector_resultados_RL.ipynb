```python
import pybullet as p
import pybullet_data as pd
p.connect(p.DIRECT)
p.setAdditionalSearchPath(pd.getDataPath())
p.loadURDF("plane.urdf")
rid = p.loadURDF("2_legged_human_like_robot12DOF_anckle.urdf", [0,0,1.2])

print("[MARK] before planar")
cid_planar = p.createConstraint(rid,-1,-1,-1, p.JOINT_PLANAR, [0,1,0], [0,0,0], [0,0,0])
p.changeConstraint(cid_planar, maxForce=1e9)
print("[MARK] after planar (OK)")

print("[MARK] before yaw-gear")
cid_yaw = p.createConstraint(rid,-1,-1,-1, p.JOINT_GEAR, [0,0,1], [0,0,0], [0,0,0])
p.changeConstraint(cid_yaw, gearRatio=0, erp=1.0, maxForce=1e9)
print("[MARK] after yaw-gear (OK)")

print("[MARK] before roll-gear")
cid_roll = p.createConstraint(rid,-1,-1,-1, p.JOINT_GEAR, [1,0,0], [0,0,0], [0,0,0])
p.changeConstraint(cid_roll, gearRatio=0, erp=1.0, maxForce=1e9)
print("[MARK] after roll-gear (OK)")

for _ in range(50):
    p.stepSimulation()

p.disconnect()
print("[DONE]")
```
